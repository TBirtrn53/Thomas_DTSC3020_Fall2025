## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue='species')
plt.title('Bill Length vs. Bill Depth by Species')
plt.show()
```

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=10)
plt.title('Histogram of Bill Length (10 bins)')
plt.show()
```

```python
sns.histplot(data=penguins, x='bill_length_mm', bins=30)
plt.title('Histogram of Bill Length (30 bins)')
plt.show()
```

## 5) Boxplot of body mass by species
