
## Part 1 – Initial Setup & Sample Data (0.5 point)

  - `Read`   (boolean: `True` if read, `False` if unread)
- Store the data in a `pandas` DataFrame.

**Hint:**


## Part 3 – Add New Book Form (1.0 point)

**Requirements:**

- Create a small form that allows the user to add a new book.
- The form must include inputs for:
  - `Title`  (text input)
  - `Author` (text input)
  - `Genre`  (selectbox or text input)
  - `Pages`  (number input, must be > 0)
  - `Read`   (checkbox)
- When the user submits the form:
  - **Validate** the inputs:
    - `Title` and `Author` cannot be empty.
    - `Pages` must be a positive number.
  - If validation passes:
    - Append the new book to the DataFrame.
    - Show a **success message**.
    - Update the displayed table.
  - If validation fails:
    - Show an **error message**.

**Hints:**
- You can use `st.form()` and `st.form_submit_button()` to create the form.
- To update the table, you must update the same DataFrame you used in Part 2.

>  Write your code for **Part 3** in the next cell.


```python
# Write your code for Part 3 here

# Get the current DataFrame from session_state
if "books_df" not in st.session_state:
    st.session_state["books_df"] = pd.DataFrame()
books_df = st.session_state["books_df"]

st.subheader("Add a New Book")

with st.form("add_book_form"):
    title = st.text_input("Title")
    author = st.text_input("Author")
    genre = st.selectbox("Genre", options=["Fiction", "Non-Fiction", "Sci-Fi", "Romance", "Dystopian", "Classic", "Fantasy", "Thriller", "Biography", "History", "Poetry", "Self-Help", "Young Adult", "Children's"], index=0)
    pages = st.number_input("Pages", min_value=1, step=1, value=1)
    read = st.checkbox("Read")

    submitted = st.form_submit_button("Add Book")

    if submitted:
        # Validation logic
        if not title:
            st.error("Title cannot be empty.")
        elif not author:
            st.error("Author cannot be empty.")
        elif pages <= 0:
            st.error("Number of pages must be positive.")
        else:
            # Append the new book to the DataFrame
            new_book = pd.DataFrame([{"Title": title, "Author": author, "Genre": genre, "Pages": pages, "Read": read}])
            st.session_state["books_df"] = pd.concat([st.session_state["books_df"], new_book], ignore_index=True)
            st.success(f"Book '{title}' added successfully!")
            # Re-run the Streamlit app to update the display (this is implicitly handled by Streamlit's rerun mechanism
            # when st.session_state is updated and the app is running in 'streamlit run' mode).
            # In Colab's direct execution, a re-execution of the whole script would be needed to see the update directly.


# Display the updated DataFrame after a potential addition
st.subheader("Current Book Library (Updated)")
st.dataframe(st.session_state["books_df"])
```
